```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
pd.read_csv('mtcars.csv')
```

```python
cars_data = pd.read_csv('mtcars.csv')
```

```python
cars_data
```

## Correlation

```python
cars_data_corr=cars_data.corr().round(2)
```

```python
cars_data_corr
```

## Range 0 - 1
-  0- 0.19- very weak 
-  0.20 -0.39- weak 
-  0.40-0.59- moderate 
-  0.60- 0.79- strong 
-  0.80- 1- very strong 

```python
plt.figure(figsize=(10,6))
sns.heatmap(data=cars_data_corr, annot=True)
plt.show()
```
